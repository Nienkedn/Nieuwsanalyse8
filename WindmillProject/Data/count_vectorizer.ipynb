```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

df = pd.read_csv('data.csv', encoding="utf-16")

df.head()
```

```python
df.Content.values
```

```python
cv = CountVectorizer()
count = cv.fit_transform(df.Content)
```

```python
cv.vocabulary_
```

```python
inv_vocabulary = {idx: word for word, idx in cv.vocabulary_.items()}
inv_vocabulary[1255]
```

```python
is_twitter = [row_num for row_num, value in enumerate(df.Origin)if value == 'twitter']
len(is_twitter), sum(is_twitter)
```

```python
words = sorted([word for word in cv.vocabulary_], key = lambda w:cv.vocabulary_[w])
df_twitter_count = pd.DataFrame(list(zip(words, np.array(count[is_twitter,:].sum(axis=0))[0])))
```
